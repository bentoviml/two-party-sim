```python
proposers = [
    ("BinarySearchProposer", lambda: BinarySearchProposer(min_offer=0, max_offer=100)),
    ("ConcedingProposer", lambda: ConcedingProposer(start_offer=90, decrement=5)),
    ("TitForTatProposer", lambda: TitForTatProposer()),
    ("RandomProposer", lambda: RandomProposer(distribution="normal", mean=50, stddev=15)),
]

responders = [
    ("UtilitarianResponder", lambda: UtilitarianResponder()),
    ("TitForTatResponder", lambda: TitForTatResponder()),
    ("ProbabilisticResponder", lambda: ProbabilisticResponder(alpha=1.0)),
]

df = run_tournament(proposers, responders, game_kwargs={
    "player_1_bad": -50,
    "player_2_bad": -100,
    "proposer_bad": -25,
    "receiver_bad": 25,
    "min_offer": 0,
    "max_offer": 100,
})
```

```python
df.head()
```

```python
summary = (
    df.groupby(["p1_proposer", "p1_responder", "p2_proposer", "p2_responder"])
    .agg(
        p1_avg_utility=("p1_utility", "mean"),
        p2_avg_utility=("p2_utility", "mean"),
        avg_rejections=("rejections", "mean"),
        n_trials=("trial", "count"),
    )
    .reset_index()
)
```

```python
summary["utility_gap"] = summary["p1_avg_utility"] - summary["p2_avg_utility"]
summary["abs_gap"] = summary["utility_gap"].abs()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Combine strategy names
summary["p1_strategy"] = summary["p1_proposer"] + " | " + summary["p1_responder"]
summary["p2_strategy"] = summary["p2_proposer"] + " | " + summary["p2_responder"]

# Pivot for heatmap
pivot = summary.pivot(index="p1_strategy", columns="p2_strategy", values="utility_gap")

# Plot
plt.figure(figsize=(14, 10))
sns.heatmap(pivot, annot=True, fmt=".0f", center=0, cmap="coolwarm")
plt.title("Utility Gap (Player 1 − Player 2)")
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# For every fixed P2 strategy, find the P1 strategy that maximizes p1_avg_utility
best_response_p1 = (
    summary.loc[summary.groupby(["p2_proposer", "p2_responder"])["p1_avg_utility"].idxmax()]
    .sort_values(["p2_proposer", "p2_responder"])
    .reset_index(drop=True)
)

best_response_p2 = (
    summary.loc[summary.groupby(["p1_proposer", "p1_responder"])["p2_avg_utility"].idxmax()]
    .sort_values(["p1_proposer", "p1_responder"])
    .reset_index(drop=True)
)
```

```python
p1_best = summary.sort_values(by="p1_avg_utility",ascending=False)
```

```python
p1_best
```

```python
p2_best = summary.sort_values(by="p2_avg_utility",ascending=False)
```

```python
p2_best
```

```python
# Merge both best response tables on all strategy fields
merged = pd.merge(
    best_response_p1,
    best_response_p2,
    how="inner",
    on=["p1_proposer", "p1_responder", "p2_proposer", "p2_responder"],
    suffixes=("_p1", "_p2")
)

# Only keep relevant columns
mutual_best = merged[[
    "p1_proposer", "p1_responder", "p2_proposer", "p2_responder",
    "p1_avg_utility_p1", "p2_avg_utility_p2", "avg_rejections_p1"
]].rename(columns={
    "p1_avg_utility_p1": "p1_avg_utility",
    "p2_avg_utility_p2": "p2_avg_utility",
    "avg_rejections_p1": "avg_rejections"
})

mutual_best = mutual_best.sort_values("p1_avg_utility", ascending=False).reset_index(drop=True)
```
